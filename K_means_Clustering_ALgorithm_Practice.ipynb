```python
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split


import pandas as pd 
import numpy as np 
```

```python
# Note we have to ignore value of Y from here as this is a clustering algorithm.
X, Y = make_blobs(n_samples = 1500, centers = 4, n_features = 2, random_state = 39)



X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.30, random_state = 35)


plt.scatter(X_train[:, 0], X_train[:, 1])
```

```python
# using elbow method to find the value of K :

wcss = []

for k in range(1, 15):
    kmeans = KMeans(n_clusters = k, init = 'k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# create the elbow curve 

plt.plot(range(1, 15), wcss)
plt.xticks(range(1, 15))
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

```python
# clearly from above we can observe that our curve saturates on k = 4 

kmeans = KMeans(n_clusters = 4, init = 'k-means++')
```

```python
y_pred = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c =y_pred)
```

```python
y_test_result = kmeans.fit_predict(X_test)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_test_result)
```
