```python
file_names = os.listdir(r'datasets/dog_vs_cat_data/train/')

dog_count = 0
cat_count = 0

for img_file in file_names:
    name = img_file[0:3]
    if name == 'dog':
        dog_count += 1
        
    else:
        cat_count += 1
        
        
        
print('Number of dog images = ', dog_count)  
print('Number of cat images = ', cat_count)


file_names[12501]
```

```python
## create a new dir for resized images 

os.mkdir(r'datasets/dog_vs_cat_data/train/image_resized')
```

```python
original_folder = r'datasets/dog_vs_cat_data/train/'
resized_folder = r'datasets/dog_vs_cat_data/train/image_resized/'


for i in range(2000):
    
    filename_cat = os.listdir(original_folder)[i] 
    filename_dog = os.listdir(original_folder)[12501 + i]
    
    cat_img_path = original_folder + filename_cat
    dog_img_path = original_folder + filename_dog
    
    cat_img = Image.open(cat_img_path)
    dog_img = Image.open(dog_img_path)
    
    cat_img = cat_img.resize((224, 224))
    cat_img = cat_img.convert('RGB')
    
    dog_img = dog_img.resize((224, 224))
    dog_img = dog_img.convert('RGB')
    
    newCatImgPath = resized_folder + filename_cat
    newDogImgPath = resized_folder + filename_dog
    
    cat_img.save(newCatImgPath)
    dog_img.save(newDogImgPath)
    


    
```

```python
## display resized dog image 

img = mpimg.imread(r'datasets/dog_vs_cat_data/train/image_resized/dog.100.jpg')
imgplt = plt.imshow(img)
plt.show()
```
