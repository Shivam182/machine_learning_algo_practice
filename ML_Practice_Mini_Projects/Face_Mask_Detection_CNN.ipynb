```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
import cv2 
import os 

from PIL import Image 
from sklearn.model_selection import train_test_split 
```

```python
mask_datapath = r"datasets/face_mask_data/data/with_mask/"
no_mask_datapath = r"datasets/face_mask_data/data/without_mask/"
```

```python
with_mask_img = os.listdir(mask_datapath)


# first 5 file names 
print(with_mask_img[0:5])


# last 5 file names 
print(with_mask_img[-5:])
```

```python
without_mask_img = os.listdir(no_mask_datapath)



# first 5 files 
print(without_mask_img[0:5])

# last 5 files 
print(without_mask_img[-5:])
```

```python
print('Number of masked images :', len(with_mask_img))
print('Number of un-masked images :', len(without_mask_img))


## data is balanced !!!
```

```python
## now we classify the images as masked & non-masked into binary 0 and 1 labels 


with_mask_labels = [1] * len(with_mask_img)

without_mask_labels = [0] * len(without_mask_img)
```

```python
print(with_mask_labels[0:5])
print(without_mask_labels[-5:])
```

```python
labels = with_mask_labels + without_mask_labels 
```

```python
print(labels[0:5])
print(labels[-5:])
```

```python
## display the images 

img = mpimg.imread('datasets/face_mask_data/data/with_mask/with_mask_2590.jpg')

imgplot = plt.imshow(img)

plt.show()
```

```python
## now we have to make the dimensions of all the images same as they are all different 

# 1. Resize the images 
# 2. Convert the images into numpy array 

data = []

for img in with_mask_img:
    image = Image.open(mask_datapath  + img ) # open method is for reading those images 
    image = image.resize((128, 128))
    image = image.convert('RGB')
    
    image = np.array(image)
    
    data.append(image)

    
    
for img in without_mask_img:
    image = Image.open(no_mask_datapath  + img ) # open method is for reading those images 
    print(no_mask_datapath+img)
    image = image.resize((128, 128))
    image = image.convert('RGB')
    
    image = np.array(image)
    
    data.append(image)
```

```python
X = np.array(data)
Y = np.array(labels)

```

```python
print(Y)
print(X.shape)
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)

print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
## print images 

plt.imshow(X_test[211])
```

```python
print(np.bincount(Y_test)) ## all the data seems to be balanced then whre is the issue ???
```

```python
## Scaling the data 

X_train_scaled = X_train / 255 
X_test_scaled = X_test / 255
```

```python
## Building CNN 
```

```python
import tensorflow as tf 
from tensorflow import keras 

```

```python
# !pip install tensorflow
```

```python
num_of_classes  = 2 

model = keras.Sequential([
    
    keras.layers.Conv2D(32, kernel_size=(3, 3), activation = 'relu', input_shape = (128, 128, 3)),
    keras.layers.MaxPooling2D(pool_size = (2, 2)),
    
    keras.layers.Conv2D(64, kernel_size=(3, 3), activation = 'relu'),
    keras.layers.MaxPooling2D(pool_size = (2, 2)),
    
    keras.layers.Flatten(),
    
    keras.layers.Dense(128, activation = 'relu'), 
    keras.layers.Dropout(0.5), 
    
    keras.layers.Dense(64, activation = 'relu'), 
    keras.layers.Dropout(0.5), 
    
    keras.layers.Dense(num_of_classes, activation = 'sigmoid')
    
])
```

```python
## compile the neural network 

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])

```

```python
## Training the neural netowrk 

history =model.fit(X_train_scaled, Y_train, validation_split = 0.1, epochs = 15 )


```

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test)

print('Test Accuracy = ', accuracy )
```

```python
h = history 

## plot the loss value 
plt.plot(h.history['loss'], label = 'train_loss')
plt.plot(h.history['val_loss'], label = 'validation_loss')
plt.legend()
plt.show()


plt.plot(h.history['acc'], label = 'train_acc')
plt.plot(h.history['val_acc'], label = 'validation_acc')
plt.legend()
plt.show()
```

```python
## Building the predictive system 
input_image_path = input('Path of the image to be predicted: ')

input_image = cv2.imread(input_image_path)

#cv2.imshow('image window',input_image)

input_image_resized = cv2.resize(input_image, (128,128))

input_image_scaled = input_image_resized/255

input_image_reshaped = np.reshape(input_image_scaled, [1,128,128,3])

input_prediction = model.predict(input_image_reshaped)

print(input_prediction)


input_pred_label = np.argmax(input_prediction)

print(input_pred_label)


if input_pred_label == 1:

  print('The person in the image is wearing a mask')

else:

  print('The person in the image is not wearing a mask')

```
